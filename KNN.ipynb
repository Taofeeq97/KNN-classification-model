```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors='coerce')
```

```python
df.info()
```

```python
df.dropna(how='any', inplace=True)
```

```python
df.info()
```

```python
df.Churn.value_counts()
```

```python
X = df.drop(['customerID', 'Churn'], axis=1)
```

```python
y=df.Churn.values
```

```python
y
```

#Convert categorical X variables to numerical to be able to plug it into ML model ---> Feature Encoding

```python
X.columns
```

```python
X = pd.get_dummies(X, columns=['gender', 'Partner', 'Dependents', 'tenure',
       'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'], drop_first=True, dtype=int)
```

```python
X
```

#Splitting to train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
len(x_train)
```

```python
x_train.head()
```

Features scaling

```python
x_test.head(1)
```

```python
x_test.columns
```

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

#KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)

y_pred = knn.predict(x_test)
y_pred
```

```python
y_test
```

# Model classification performance metrics

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
accuracy_score(y_test, y_pred)
```
